# Missing Value

```python
url_titanic = '/content/drive/MyDrive/Colab Notebooks/MSIB/Pengenalan Python untuk Data Science/PDA_DataSource/Pengolahan_Data/titanic.csv'
titanic = pd.read_csv(url_titanic)
titanic.head()
```

## cek misising value

```python
titanic.isnull().sum()
```

## Imputing Missing Value

```python
titanic[titanic['Embarked'].isnull()]
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Embarked'] = titanic['Embarked'].fillna('S')
```

```python
titanic[titanic['PassengerId'] == 62]
```

```python
titanic.isnull().sum()
```

```python
# Inputation with mean or median
# dapat dilihat drai histogram
titanic.hist('Age')
```

- dilihat dari distribusinya
- normal atau skewed (miring)
- skewed = median (karena apabila pakai mean akan menggagu nilai distribusi)
- normal = mean
- central tendency

```python
median_age = titanic['Age'].median()
median_age
```

```python
titanic['Age'] = titanic['Age'].fillna(median_age)
```

```python
titanic.isnull().sum()
```

```python
def fill_missing_by_pclass(dataset, column):
  for i in range(1,4):
    median = dataset[dataset['Pclass'] == i][column].median()
    dataset[column] = dataset[column].fillna(median)
    return dataset
```

```python
fill_missing_by_pclass(titanic, 'Age')
```
